# Face Detection on Person's image Using OpenCV

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading trained XML file for detecting a face


```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

### Function to detect a face and draw a rectangle around it

```python
def find_face(cascade, img):
  
    # Converting image into gray-scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detecting Face
    faces = cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=2);
    
    # Drawing a Blue Rectangle around the face
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
    return img
```

### Testing

```python
#load another image
pic = cv2.imread('pic1.jpg')

#call our function to detect faces
img = find_face(face_cascade, pic)

#conver image to RGB and show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
